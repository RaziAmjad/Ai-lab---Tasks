```python
class FizzBuzz:
    def __init__(self, start, end):
        self.start = start
        self.end = end

    def play(self):
        for i in range(self.start, self.end + 1):
            if i % 3 == 0 and i % 5 == 0:
                print("Fizz Buzz")
            elif i % 3 == 0:
                print("Fizz")
            elif i % 5 == 0:
                print("Buzz")
            else:
                print(i)

game = FizzBuzz(1, 100)
game.play()
```

```python
class MovieBudgetAnalyzer:
    def __init__(self, movies):
        self.movies = movies

    def calculate_average_budget(self):
        total_budget = sum(budget for _, budget in self.movies)
        return total_budget / len(self.movies)

    def show_high_budget_movies(self):
        average = self.calculate_average_budget()
        print(f"\nAverage Movie Budget: {average:,}\n")
        
        count = 0
        print("Movies with Higher than Average Budget:\n")
        for name, budget in self.movies:
            if budget > average:
                difference = budget - average
                print(f"{name} — {budget:,} (↑ {difference:,} ")
                count += 1
        print(f"\nTotal movies above average budget: {count}")

    def add_movies(self):
        num = int(input("How many movies do you want to add? "))
        for _ in range(num):
            name = input("Enter movie name: ")
            budget = int(input("Enter movie budget "))
            self.movies.append((name, budget))
        print("\nMovies added successfully!\n")


movies = [
    ("Khuda Kay Liye", 20000000),
    ("Bol", 9000000),
    ("Waar", 4500000),
    ("Jumanji", 379000000),
    ("Pony tales", 365000000),
    ("The Legend of Maula Jatt", 356000000),
    ("Parwaaz Hai Junoon", 200000000)
]

analyzer = MovieBudgetAnalyzer(movies)
analyzer.show_high_budget_movies()
```
